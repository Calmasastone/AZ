```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('United_Data.xlsx')
```

```python
df.head(3)
```

```python
# checking all the rows with nulls
df[df.isna().any(axis=1)]
```

```python
Old_Salary = df['SALARY'].sum()
```

```python
New_Salary = df['NEW_SALARY'].sum()
```

```python
Dif_Per = round((New_Salary/Old_Salary-1)*100, 1)
```

```python
# let's check how salary grew up in %
Dif_Per
```

```python
# making the column with new salary
Compare_Salary = df.groupby(['DEPARTMENT_NAME'], as_index=False)\
.aggregate({'SALARY': 'sum', 'NEW_SALARY': 'sum'})\
.sort_values('NEW_SALARY', ascending=False)
```

```python
Compare_Salary
```

```python
# making a copy for safe following calculations (to not to break clear df)
Test = Compare_Salary
```

```python
# cheking the difference between old and new salaries in departments in percent
Test['PERCENT'] = round((Test['NEW_SALARY']/Test['SALARY']-1)*100, 1)
```

```python
Test
```

```python
# adding 'difference' column at 3rd id
Test.insert(3, 'DIFFERENCE', Test['NEW_SALARY'] - Test['SALARY'])
```

```python
# renaming for better look
Test = Test.rename(columns={
    'DEPARTMENT_NAME': 'DEPARTMENT'
})
```

```python
# all fine, let's assign Test as main df
Compare_Salary = Test
```

```python
Compare_Salary
```

```python
# let's calculate work experience of our employees
Work_Experience_Years = df.groupby(['EXPERIENCE'], as_index=False)\
.aggregate({'EXPERIENCE': 'count'})
```

```python
Work_Experience_Years
```
